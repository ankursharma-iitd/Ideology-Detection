```python
def check_policy(sent, policies):
    p = []
    words = []
    for policy, keywords in policies.items():
        for word in keywords:
            if is_phrase_in(word, sent):
                words.append(word)
                p.append(policy)
                break
    if len(p)>1:
        print(p)
    return p, words
```

```python
output = open('./new-out-all-by-stmnts.txt','w')
for entity in dicti:
    for source in dicti[entity]:
        for line in dicti[entity][source]:
            p, words = check_policy(line, policies)
            p = " ".join(p)
            words = " ".join(words)
            output.write(entity + ',' + source + ',' + p + " -- " + words + "\n")
output.close()
```

```python
econ_lines = dict()
tech_lines = dict()
with open("./new-out-all-by-stmnts.txt",'r') as file_o: 
    lineList = file_o.readlines()
    count = 0
    for entity in dicti:
        econ_lines[entity] = {}
        tech_lines[entity] = {}
        for source in dicti[entity]:
            econ_lines[entity][source] = []
            tech_lines[entity][source] = []
            for line in dicti[entity][source]:
                y = lineList[count]
                y_list = y.split(',')
                y = y_list[2].strip().split(" ")
                if len(y)>1:
                    print(y)
                assert y_list[0] == entity and y_list[1] == source
                for policy in y:
                    if policy=="econ":
                        econ_lines[entity][source] += [line]
                    if policy=="tech":
                        tech_lines[entity][source] += [line]
                count += 1
```
